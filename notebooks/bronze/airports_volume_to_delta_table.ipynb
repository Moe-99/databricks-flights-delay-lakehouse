```python
from pyspark.sql.functions import current_timestamp, lit
volume_path = "/Volumes/dev_project/flight_delay_lakehouse/raw_files/airports.csv"
airports_df = (spark.read.csv(volume_path, header=True, inferSchema=True)
    .withColumn("ingested_at", current_timestamp())
    .withColumn("source", lit("airports.csv"))
)

(airports_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("dev_project.flight_delay_lakehouse.bronze_airports_raw")
)
```
