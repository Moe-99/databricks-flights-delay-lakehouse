```python
import requests 
csv_url = "https://ourairports.com/data/airports.csv"
volume_path = "/Volumes/dev_project/flight_delay_lakehouse/raw_files/airports.csv"
r = requests.get(csv_url, timeout = 60)
print("download status", r.status_code)
print("bytes downloaded", len(r.content))
with open(volume_path, "wb") as f:
    f.write(r.content)
print("file saved to", volume_path)


df = spark.read.csv(volume_path, header=True, inferSchema=True)
print("Row count:", df.count())

df.select("iata_code", "name", "latitude_deg", "longitude_deg", "iso_country") \
  .where("iata_code is not null") \
  .show(10, truncate=False)

df.where("iata_code in ('AMS','LHR','JFK')") \
  .select("iata_code", "name", "latitude_deg", "longitude_deg", "iso_country") \
  .show(truncate=False)

```
