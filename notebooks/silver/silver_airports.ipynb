```python
from pyspark.sql.functions import col, trim
#Reading airports raw table to a dataframe
bronze_airports_df = spark.table("dev_project.flight_delay_lakehouse.bronze_airports_raw")

#Constructing the airports silver table with some cleaning work
silver_airports_df = bronze_airports_df.select(
    trim(col("iata_code")).alias("airport_iata"),
    trim(col("name")).alias("airport_name"),
    trim(col("type")).alias("airport_type"),
    trim(col("iso_country")).alias("country_code"),
    trim(col("municipality")).alias("municipality"),
    trim(col("icao_code")).alias("airports_icao"),

    col("latitude_deg").alias("airport_latitude"),
    col("longitude_deg").alias("airport_longitude"),
    col("ingested_at"),
    col("source")
    ).where(col("iata_code").isNotNull())

(
    silver_airports_df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable("dev_project.flight_delay_lakehouse.silver_airports")
    )

```
