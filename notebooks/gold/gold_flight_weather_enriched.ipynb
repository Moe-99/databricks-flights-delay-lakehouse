```sql
CREATE OR REPLACE TABLE dev_project.flight_delay_lakehouse.gold_flight_weather_enriched AS
SELECT
  f.flight_iata,
  f.flight_number,
  f.airline_name,
  f.airline_iata,
  f.flight_status,
  f.dep_airport_iata,
  f.arr_airport_iata,
  f.dep_scheduled_ts,
  date_trunc('hour', f.dep_scheduled_ts) AS dep_hour_ts,
  f.dep_delay_minutes,

  w.air_temperature_2m_c,
  w.precipitation_mm,
  w.wind_speed_10m_kmh,
  w.weather_kind,

  f.ingested_at
FROM dev_project.flight_delay_lakehouse.silver_flights f
LEFT JOIN dev_project.flight_delay_lakehouse.silver_weather_hourly w
ON f.dep_airport_iata = w.airport_iata
AND date_trunc('hour', f.dep_scheduled_ts) = w.weather_ts


```

```sql
SELECT * FROM dev_project.flight_delay_lakehouse.gold_flight_weather_enriched
```
