```sql
CREATE OR REPLACE TABLE dev_project.flight_delay_lakehouse.gold_airport_delay_summary AS 
SELECT
  f. dep_airport_iata,
  a.airport_name,
  a.country_code,
  CAST(dep_scheduled_ts AS DATE) AS flight_date,
  COUNT(*) AS total_flights,
  SUM(CASE WHEN dep_delay_minutes > 15 THEN 1 ELSE 0 END) AS delayed_flights,
  round(AVG(dep_delay_minutes), 2) AS avg_dep_delay_minutes,
  round(100.0 * SUM(CASE WHEN dep_delay_minutes IS NULL OR dep_delay_minutes <= 15 THEN 1 ELSE 0 END) / COUNT(*), 2) AS on_time_rate,
  round((delayed_flights * 100.0) / total_flights , 2) AS delay_rate
FROM dev_project.flight_delay_lakehouse.silver_flights f 
LEFT JOIN dev_project.flight_delay_lakehouse.silver_airports a 
ON f.dep_airport_iata = a.airport_iata
GROUP BY 
  dep_airport_iata,
  CAST(dep_scheduled_ts AS DATE),
  a.airport_name,
  a.country_code




```
